```python
def empty_cell(dataframe, column:str):
    count_nan_column = dataframe[column].isna().sum()
    print(f"Empty Rows Count of {column}: {count_nan_column}")

    dataframe = dataframe.dropna(subset=[column])
    return dataframe
```
